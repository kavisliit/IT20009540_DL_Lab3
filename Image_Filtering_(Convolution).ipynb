```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

### Load the image using OpenCV

```python
im = cv2.imread('/content/sample_data/lenna.png') #change the path as needed
```

```python
im.shape
```

### Convert the RGB image into a grayscale image

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

### Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
h_kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

### Do the filtering

```python
v_image = cv2.filter2D(im, -1, v_kernel)
h_image = cv2.filter2D(im, -1, h_kernel)
```

### Display results

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')
```

```python
sobel_x = cv2.Sobel(im, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(im, cv2.CV_64F, 0, 1, ksize=3)

plt.figure()
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X Edge')

plt.figure()
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y Edge')
```

```python
laplacian = cv2.Laplacian(im, cv2.CV_64F)

plt.figure()
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Edge')
```

```python
canny = cv2.Canny(im, threshold1=100, threshold2=200)

plt.figure()
plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detection')
```

```python
scharr_x = cv2.Scharr(im, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(im, cv2.CV_64F, 0, 1)

plt.figure()
plt.imshow(scharr_x, cmap='gray')
plt.title('Scharr X Edge')

plt.figure()
plt.imshow(scharr_y, cmap='gray')
plt.title('Scharr Y Edge')
```
